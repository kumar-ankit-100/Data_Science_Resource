```python
def size_to_int(item):
    if item[-1] == 'M':
        item = item.replace('M','')
        item = float(item) * 1024
    elif item[-1] == 'K':
        item =float( item.replace('K',''))
    else:
        item =  np.nan
    return item  
```

```python
df.Size=df['Size'].apply(size_to_int)
```
